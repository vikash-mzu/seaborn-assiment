```python
Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
Line Plot (sns.lineplot):

Use: Used to plot continuous data points (numeric data) and visualize trends over time or other continuous variables.
Example: Plotting time series data like stock prices or temperature variations.
Bar Plot (sns.barplot):

Use: Used to visualize categorical data with respect to a numerical variable by representing data with rectangular bars.
Example: Comparing sales revenue by product category or profit by company.
Scatter Plot (sns.scatterplot):

Use: Used to display the relationship between two numerical variables by plotting data points on a Cartesian plane.
Example: Visualizing the correlation between height and weight in a population.
Box Plot (sns.boxplot):

Use: Used to display the distribution of data based on five summary statistics (minimum, first quartile, median, third quartile, and maximum).
Example: Visualizing the distribution of salaries across different job positions.
Heatmap (sns.heatmap):

Use: Used to visualize matrix-like data using color shades, where each cell represents a data point and the color intensity shows its magnitude.
Example: Correlation matrices or visualizing data in a table format (like flight delays by month and year).




Que 2: Load the "fmri" dataset and plot a line plot using Seaborn.

import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Plot a line plot with timepoint on x-axis and signal on y-axis, differentiated by event and region
sns.lineplot(data=fmri, x='timepoint', y='signal', hue='event', style='region')

# Add title and show plot
plt.title('FMRI Signal Over Time')
plt.show()



Que 3: Load the "titanic" dataset and plot two box plots.

import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset
titanic = sns.load_dataset("titanic")

# Create a box plot for 'age' vs 'pclass'
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic)
plt.title('Age Distribution by Pclass')

# Create a box plot for 'fare' vs 'pclass'
plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic)
plt.title('Fare Distribution by Pclass')

# Show plot
plt.tight_layout()
plt.show()




Que 4: Use the diamonds dataset to plot a histogram for the price' column.

import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Plot a histogram for 'price' with hue based on the 'cut' column
sns.histplot(data=diamonds, x='price', hue='cut', kde=True)

# Add title and show plot
plt.title('Distribution of Diamond Prices by Cut')
plt.show()



Que 5: Use the iris dataset to plot a pair plot with hue for the "species column.

import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset("iris")

# Plot a pair plot with hue for the 'species' column
sns.pairplot(iris, hue='species')

# Show plot
plt.show()



Que 6: Use the "flights dataset to plot a heatmap.

import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights = sns.load_dataset("flights")

# Create a pivot table for the heatmap
flights_pivot = flights.pivot("month", "year", "passengers")

# Plot a heatmap
sns.heatmap(flights_pivot, annot=True, fmt='d', cmap='Blues')

# Add title and show plot
plt.title('Number of Passengers per Month and Year')
plt.show()
```
